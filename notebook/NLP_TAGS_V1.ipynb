```python

# Extract company name and model no. both.

# Function to extract two pattern tags                
def match2(token_pos,pos1,pos2):
    for subsen in token_pos:
        # avoid index error and catch last three elements
        end = len(subsen) - 1
        for ind, (a, b) in enumerate(subsen, 1):
            if ind == end:
                break
            if b == pos1 and subsen[ind][1] == pos2:
                yield ("{} {}".format(a, subsen[ind][0], subsen[ind + 1][0]))

# Print company and model no for each sentence

for i in range(0,3):
    tokens = word_tokenize(new_comments[i]) # Generate list of tokens
    tokens_pos = pos_tag(tokens)
    a = [tokens_pos]
    print(list(match2(a,'NNP','NN')))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def extract_company_name(comments):
    company_names = []
    for comment in comments:
        tokenized_comment = word_tokenize(comment)
        tagged_comment = pos_tag(tokenized_comment)
        for word, tag in tagged_comment:
            if tag == 'NNP' and word.isalpha():
                company_names.append(word)
    return company_names

def replace_spaces(comments):
    return [comment.replace("_", " ").replace("-", " ") for comment in comments]

comments = ["01-MIC_G36 & 5.56 S4HANA_SCM_MOM_09062021",
            "01-MIC_G36 & 5.56 S4HANA_SCM_MOM_09062021",
            "01-MIC_G36 & 5.56 S4HANA_WSR_MOM_28042021",
            "02-MIC-G36 & 5.56 ERP Project-WSR V1.0-06052021"
           ]

new_comments = replace_spaces(comments)
company_names = extract_company_name(new_comments)
print(company_names)
```

```python
import os

def remove_file_extensions(filenames):
    return [os.path.splitext(filename)[0] for filename in filenames]

def replace_spaces(comments):
    return [comment.replace("_", " ").replace("-", " ") for comment in comments]


newfilenames = remove_file_extensions(filenames)
new_filenames = replace_spaces(newfilenames)
new_filenames
```

```python
import csv
from nltk import word_tokenize, pos_tag

# Open a CSV file for writing
with open('output.csv', 'w', newline='') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)

    # Write the header row
    csv_writer.writerow(['fileName', 'tags'])

    # Loop through the new_filenames
    for i in range(0,len(new_filenames)):
        token_comment = word_tokenize(new_filenames[i])
        tagged_comment = pos_tag(token_comment)

        # Get the NNP words
        nnp_words = [word for word, tag in tagged_comment if (tag=='NNP') and word.isalpha()]

        # Write the filename and the NNP words to the CSV file
        csv_writer.writerow([new_filenames[i], ', '.join(nnp_words)])
```
